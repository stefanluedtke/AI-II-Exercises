## Pandas
Ein weiteres Paket, das wir häufig verwenden werden, ist `pandas`. Dieses führt das Data Frame als zentrale Datenstruktur ein. Ein Data Frame ist eine Tabelle, wobei einzelne Spalten über Spaltennamen angesprochen werden können. Pandas bietet viele nützliche Möglichkeiten, auf Data Frames zuzugreifen. 

```python
import numpy as np
import pandas as pd
from scipy.io import arff
import matplotlib.pyplot as plt

data = arff.loadarff('S08.arff')
df = pd.DataFrame(data[0])


#plt.plot(df["Sensor_T8_Acceleration_X"])
df
```

### Auswahl von Zeilen / Spalten
Mit pandas können wir leicht bestimmte Zeilen oder Spalten in einem Data Frame auswählen. Eine Möglichkeit ist die Auswahl über den Namen der Spalte:

```python
accx = df.loc[:,"Sensor_T8_Acceleration_X"]
```

Hierbei steht `:` für "wähle alle Zeilen aus". Wenn nur eine einzelne Spalte zurückgegeben wird, ist das Ergebnis vom Typ `Series`, ansonsten ist das Ergebnis wieder ein Data Frame. Spalten können auch über den Index zugegriffen werden:

```python
acc = df.iloc[:,1:4]
acc
```

Äquivalent kann auf die Zeilen zugegriffen werden. Der folgende Ausdruck liefert beispielsweise nur die ersten 5 Zeilen:

```python
df.iloc[0:5,:]
```

Natürlich können wir auch beides kombinieren, z.B. so:

```python
df.loc[0:5,["Sensor_T8_Acceleration_X", "Sensor_T8_Acceleration_Y"]]
```

Eine weitere Möglichkeit ist, Zeilen (oder Spalten) über bool'sche Ausdrücke zuzugreifen. Beispielsweise liefert folgender Ausdruck alle Zeilen, in denen der Wert von `Sensor_T8_Acceleration_X` kleiner als 0.7 ist:

```python
df.loc[df.Sensor_T8_Acceleration_X < 0.7,:]
```

### Einfügen neuer Werte
Das schreiben neuer Werte in ein Data Frame ist ebenso leicht möglich. Der folgende Ausdruck setzt alle Werte der Spalte `Sensor_T8_Acceleration_Y`, die kleiner als 0 sind, auf den Wert 0.

```python
df.loc[df.Sensor_T8_Acceleration_Y < 0,"Sensor_T8_Acceleration_Y"] = 0
df
```

### Apply
Oftmals möchten wir Funktionen Zeilen- oder Spaltenweise auf Daten anwenden. Die Funktion apply ermöglicht das auf einfache Art. Der folgende Ausdruck berechnet die Mittelwerte pro Spalte:

```python
df.iloc[:,1:31].apply(np.mean)
```

## Aufgabe 2
Als nächstes wollen wir Features für die Daten berechnen. Für sequentielle Daten werden die Features typischerweise segmentiert berechnet, d.h. es wird für die Werte 1 bis n ein Feature (z.B. Mittelwert, ...) berechnet, dann für die Werte n+1 bis 2n usw. Die Fenster können sich dabei auch überlappen. In der Vorlesung werden Sie später
noch weitere Features kennenlernen, z.B. Features, die sich auf die Frequenz des Signals beziehen.

Implementieren Sie eine Funktion feature, die mit Angabe einer Fenstergröße, einer Überlappungslänge, eines Datensatzes und einer statistischen Merkmalsfunktion (mean,...) das jeweilige Merkmal auf den gegebenen Fenstern berechnet. Berechnen sie den Mittelwert, den Median und die Varianz der Beschleunigungsdaten am rechten Bein mit den Fenstergrößen 128, 256 und 512. Verwenden Sie dabei halbüberlappende Fenster und plotten sie das Ergebnis.

```python
import matplotlib.pyplot as plt

def feature(wsize,overlap,data,ffunct):
    start = 0
    numwindows = int(len(data)/(wsize-overlap))
    result = np.zeros(numwindows)
    for i in range(numwindows):
        end = start + wsize-1
        windowed = data[start:end]
        result[i] = ffunct(windowed)
        start = start + wsize - overlap
    return(result)

meanX = feature(64,32,df.loc[:,"Sensor_T8_Acceleration_X"].values,np.var)
plt.plot(meanX)
```
